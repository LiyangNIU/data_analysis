```python
# 导入pandas模块，简称为pd
import pandas as pd
# 读取文件，并设置参数encoding = "utf-8"
df = pd.read_csv("snack.csv",encoding = "utf-8")

# 创建空列表df_list用于存储物品
df_list = []

# for循环遍历df["物品"]
for i in df["物品"]:
    
    # 用split()函数将i拆分为列表，并将返回的列表赋值给temp
    temp=i.split(",")
    
    # 再次for循环遍历temp，获取单个物品
    for j in temp:
        
        # TODO 用append()函数将j，添加到空列表df_list中
        df_list.append(j)
        
# 使用pd.Series()将列表df_list转成Series，赋值给goods
goods=pd.Series(df_list)

# 使用value_counts()统计goods各个物品的被购买次数，赋值给goods_counts
goods_counts=goods.value_counts()

# 导入matplotlib.pyplot，并使用"plt"作为该模块的简写
import matplotlib.pyplot as plt
# 设置显示中文字体"Arial Unicode MS"
plt.rcParams["font.sans-serif"] = "Arial Unicode MS"

# 使用plt.bar()绘制柱状图
# 以goods_counts.index为x轴数据，goods_counts.values为y轴数据，设置width=0.5
plt.bar(goods_counts.index,goods_counts.values,width=0.5)

# plt.title()设置图标题为"不同物品被购买的次数"
plt.title("不同物品被购买的次数")

# plt.show()显示图像
plt.show()
```
