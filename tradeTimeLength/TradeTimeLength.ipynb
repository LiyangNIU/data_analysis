```python
import matplotlib.pyplot as plt

import pandas as pd

data = pd.read_csv("视频会员订单数据.csv")

data["create_time"]=pd.to_datetime(data["create_time"])

data["pay_time"]=pd.to_datetime(data["pay_time"])

data["time_gap"]=data["pay_time"] - data["create_time"]

data["time_gap"]=data["time_gap"].dt.days

new_data=data[(data["time_gap"]>0)&(data["time_gap"]<7)]

result=new_data["time_gap"].value_counts()


plt.bar(result.index,result.values,width=0.5)

plt.show()
```
