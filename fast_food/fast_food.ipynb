```python
import pandas as pd
df = pd.read_csv("fast_food.csv")
```

```python
food_type=df.columns[1:6]
```

```python
supports=[]

for i in food_type:
    order=df[i].value_counts()
    order=order[1]
    sup=order/100
    supports.append(sup)
print(supports)
```

```python


import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = "Arial Unicode MS"


plt.bar(food_type,supports)


plt.title("不同食品的支持率")


plt.show()
```
