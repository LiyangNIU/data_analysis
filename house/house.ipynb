```python
import pandas as pd 
df = pd.read_csv("成都市夜曲区房源数据.csv")
df=df.dropna()

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
x = df[["建筑面积"]]
y = df[['房屋总价/万']]
lr.fit(x,y)

b = lr.coef_[0][0]

a= lr.intercept_[0]

print(f"该线性回归模型为：y={b}*x+{a}")

price = lr.predict([[130]])[0][0]

print(f"房屋面积为130平方的总价为{price}万")

import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = "Arial Unicode MS"

plt.scatter(df["建筑面积"],df["房屋总价/万"],color="skyblue",label="原始数据")


plt.plot(df["建筑面积"],b*df["建筑面积"]+a, color="red",label="线性回归模型")

plt.legend()

plt.show()
```
